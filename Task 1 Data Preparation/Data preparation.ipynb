```python
import pandas as pd
df = pd.read_csv("C:/Users/Shan Sajeer/Documents/Saiket Systems Data science Internship/Customer Churn Analysis and Prediction Project/Data/Telco_Customer_Churn_Dataset  (3).csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.replace({
    "No internet service" : "No",
    "No phone service" : "No"
})
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['Churn_flag'] = df['Churn'].map({"Yes" : 1, "No" : 0})
```

```python
df['Churn_flag'].value_counts()
```

```python
cat_cols = df.select_dtypes(include = ['object']).columns.tolist()
```

```python
cat_cols = [c for c in cat_cols if c not in ['customerID','Churn']]
```

```python
df_encoded = pd.get_dummies(df, columns = cat_cols, drop_first = True)
```

```python
df_encoded.head()
```

```python
x = df_encoded.drop(columns = ['Churn_flag','Churn'],errors = 'ignore')
y = df_encoded['Churn_flag']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(
    x, y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
train_df = x_train.copy()
train_df['Churn_flag']=y_train
```

```python
test_df = x_test.copy()
test_df['Churn_flag']=y_test
```
