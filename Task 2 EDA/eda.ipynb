```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/Shan Sajeer/Documents/Saiket Systems Data science Internship/Customer Churn Analysis and Prediction Project/Data/Telco_Customer_Churn_Dataset  (3).csv")
df.head()
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title("Customer Distribution by Gender")
plt.show()
```

```python
sns.countplot(x='Partner', data=df)
plt.title("Customer Distribution by Partner")
plt.show()
```

```python
sns.countplot(x='Dependents', data=df)
plt.title("Customer Distribution by Dependents")
plt.show()
```

```python
df['tenure'].hist(bins=30, figsize=(10,5))
plt.title("Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
df['Churn_numeric'] = df['Churn'].map({'Yes': 1, 'No': 0})

contract_churn = df.groupby('Contract')['Churn_numeric'].mean() * 100
contract_churn


```

```python
contract_churn.plot(kind='bar', figsize=(8,4))
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate (%)")
plt.show()

```

```python
payment_churn = df.groupby('PaymentMethod')['Churn_numeric'].mean() * 100
payment_churn
```

```python
payment_churn.plot(kind='bar', figsize=(10,4))
plt.title("Churn Rate by Payment Method")
plt.ylabel("Churn Rate (%)")
plt.show()
```
